# Exercise 1

Consider the following LP problem $(P)$:
$$
    \begin{align} \\
        \text{min }  &15x_1+12x_2 \\
        \text{subject to } & x_1 + 2x_2 \geq 3 \\
         & -2x_1 + 4x_2 \geq -6 \\
         & x_1, x_2\geq 0 
    \end{align}
$$

Answer the following questions.

1. Find the optimal solution of $(P)$ using the graphical interpretation of the simplex method.

2. Code $(P)$ in Julia and solve it to verify that you have indeed found the optimal solution.

3. Find the dual problem of $(P)$ and denote it by $(D)$.

4. Find the optimal solution of $(D)$ using the graphical interpretation of the simplex method.

5. Code $(D)$ in Julia and solve it to verify that you have indeed found the optimal solution.
